```
import pandas as pd
import matplotlib.pyplot as plt
```

```
ena_kraken = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/graficos_arima_vs_kraken/kraken_files/ENA_VE_mensal.csv')
ena_arima = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/graficos_arima_vs_kraken/arima_files/ENA_VE_mensal.csv')
```

```
SE_ena_kraken = ena_kraken.query("lugar=='SE'")
S_ena_kraken = ena_kraken.query("lugar=='S'")
NE_ena_kraken = ena_kraken.query("lugar=='NE'")
N_ena_kraken = ena_kraken.query("lugar=='N'")
```

```
SE_ena_arima = ena_arima.query("lugar=='SE'")
S_ena_arima = ena_arima.query("lugar=='S'")
NE_ena_arima = ena_arima.query("lugar=='NE'")
N_ena_arima = ena_arima.query("lugar=='N'")
```

```
#VARIAVEL X:
data = SE_ena_kraken['data']
df = pd.DataFrame(data)
x = pd.to_datetime(df['data'])
```

```
y_SE_arima = SE_ena_arima['pMLT']
y_S_arima = S_ena_arima['pMLT']
y_NE_arima = NE_ena_arima['pMLT']
y_N_arima = N_ena_arima['pMLT']
```

```
#PLOT SE:
fig = plt.figure(figsize=(10.80,6.00), dpi=100)
plt.plot(x, SE_ena_arima['pMLT'], '-o', label = 'SE - ARIMA', color = 'blue', linewidth=3)
plt.plot(x, SE_ena_kraken['pMLT'], '-d', label = 'SE - KRAKEN', color = 'orange', linewidth=3)
plt.gca().yaxis.grid(color='gray', linestyle='-')
plt.title('SE/CO: Comparação entre os resultados ARIMA x Kraken\n Rodada 11/12/2020 - REV2')
plt.legend(fontsize='small',loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=5)
plt.ylim(40,110)
plt.axhline(y=100, linestyle='--')
plt.ylabel('% MLT')
plt.show()
```

```
#PLOT S:
fig = plt.figure(figsize=(10.80,6.00), dpi=100)
plt.plot(x, S_ena_arima['pMLT'], '-o', label = 'S - ARIMA', color = 'blue', linewidth=3)
plt.plot(x, S_ena_kraken['pMLT'], '-d', label = 'S - KRAKEN', color = 'orange', linewidth=3)
plt.gca().yaxis.grid(color='gray', linestyle='-')
plt.title('S: Comparação entre os resultados ARIMA x Kraken\n Rodada 11/12/2020 - REV2')
plt.legend(fontsize='small',loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=5)
plt.ylim(50,170)
plt.axhline(y=100, linestyle='--')
plt.ylabel('% MLT')
plt.show()
```

```
#PLOT NE:
fig = plt.figure(figsize=(10.80,6.00), dpi=100)
plt.plot(x, NE_ena_arima['pMLT'], '-o', label = 'NE - ARIMA', color = 'blue', linewidth=3)
plt.plot(x, NE_ena_kraken['pMLT'], '-d', label = 'NE - KRAKEN', color = 'orange', linewidth=3)
plt.gca().yaxis.grid(color='gray', linestyle='-')
plt.title('NE: Comparação entre os resultados ARIMA x Kraken\n Rodada 11/12/2020 - REV2')
plt.legend(fontsize='small',loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=5)
plt.ylim(30,110)
plt.axhline(y=100, linestyle='--')
plt.ylabel('% MLT')
plt.show()
```
